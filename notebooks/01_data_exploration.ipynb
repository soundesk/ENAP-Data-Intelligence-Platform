```python
import pandas as pd
import os
```

```python
base = r'C:\Users\dell\Desktop\ENAP - projet pluridisciplinaire\data\raw'
tables = ['facture', 'ligfac', 'client', 'article', 'stock', 'gamme', 'famille']
```

```python
for name in tables:
    path = os.path.join(base, f'{name}.csv')
    df = pd.read_csv(path)
    
    print(f"\n{'='*50}")
    print(f"📌 {name.upper()}")
    print(f"   Rows: {len(df)} | Columns: {len(df.columns)}")
    
    # Missing values
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print(f"   ⚠️  Missing values:")
        for col, cnt in missing.items():
            pct = (cnt/len(df))*100
            print(f"      - {col}: {cnt} ({pct:.1f}%)")
    else:
        print(f"   ✅ No missing values")
    
    # Duplicates
    dups = df.duplicated().sum()
    if dups > 0:
        print(f"   ⚠️  Duplicates: {dups}")
    else:
        print(f"   ✅ No duplicates")
```
